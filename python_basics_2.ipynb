```python
#indexing--retriving single element
#2 types--positive indexing and negative indexing
a='hello world'
print(a[1]) #positive
print(a[-10]) #negative
print(a[0])
```

```python
#slicing--retriving multiple elements
#2 types--positive slicing and negative slicing
#start:stop:step size(not mandatory)
a='hello world'
print(a[0:5])
print(a[:5])
print(a[6:11])
print(a[6:])
print(a[6:9])
print(a[-11:-6])
print(a[-5:])
```

```python
b='good morning world'
print(b[0:4:3])
print(b[5:12:2])
print(b[13:])
print(b[17:12:-1])
print(b[-1:-6:-2])
print(b[-7:-14:-1])
print(b[-7:-14:-2])
print(b[-1::-1])
```
