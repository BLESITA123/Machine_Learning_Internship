```python
import numpy as np
print("version",np.__version__)
```

```python
#creating array using list
a=[1,2,3,4,5]
print(a)
print(type(a))
x=np.array(a)
print(x)
print(type(x))
```

```python
a=[1,2,3,4,5]
print(a*2)

x=np.array([1,2,3,4,5])
print(x*2)
```

```python
#Methodds of Creating arrays

#1 zeros array

print(np.zeros(10))

#1 array of ones
print(np.ones(5))

#3 using arrange function
print(np.arange(1,10))

#arange with step size
print(np.arange(1,11,2))
print(np.arange(50,19,-6))

#Evenly spaced values
print(np.linspace(1,10,4))
print(np.linspace(1,10,5)) #1+10/2 ,1+5.5/2
```

```python
#types of array---1D, 2D, 3D
#creating 1D array
a=np.array([1,2,3,4,5,6])
print("1D array: ",a)
print("Dimension: ",a.ndim)

#creating 2D array
b=np.array([[1,2],[3,4]])
print("2D array: ",b)
print("Dimension: ",b.ndim)

#creating 3D array
c=np.array([[[1,2],[3,4]],[[4,5],[7,8]]])
print("3D array: ",c)
print("Dimension: ",c.ndim)
```

```python
#Shape,Size , Dimension and Datatype
a1=np.array([1,2,3,4,5,6,7])
print("Shape",a1.shape) #rows,columns
print("size",a1.size) #total elements
print("Dimensions:",a1.ndim)
print("Data type of values:",a1.dtype)
print(type(a1))
```

```python
#Shape,Size , Dimension and Datatype
a2=np.array([[1,2],[3,4],[3,6],[8,4]])
print("Shape",a2.shape) #rows,columns
print("size",a2.size) #total elements
print("Dimensions:",a2.ndim)
print("Data type of values:",a2.dtype)
print(type(a2))
```

```python
#Shape,Size , Dimension and Datatype
a3=np.array([[[1,2],[3,4]],[[4,5],[7,8]]])
print("Shape",a3.shape) #rows,columns
print("size",a3.size) #total elements
print("Dimensions:",a3.ndim)
print("Data type of values:",a3.dtype)
print(type(a3))
```

```python
#Aggreagate functions 
a=np.array([1,2,3,4,5,6,7,8])
print("minimum:",np.min(a)) #a.min()
print("maximum:",np.max(a))
print("mean:",np.mean(a))
print("Median:",np.median(a))
print("sum:",np.sum(a))
print("standard deviation:",np.std(a))
```
