### Lists

```python

#collection of elements,where elements can be of diff data types
#ordered collection (can be indexed and sliced)
#can be changed, can have duplicates
lst=[1,23.89,True,"hello",1,1,1]
print(lst)
print(type(lst))
print(len(lst))
#changing the element
lst[3]="hi"
print(lst)
```

### Tuple

```python

# collections of elements,where of diff data types can be stored in tuple
# ordered collection (can be indexed and sliced)
# cannot be changed
# can have duplicates
tpl=(3,24.5,False,"hello","hello")
print(tpl)
print(type(tpl))
print(len(tpl))
#tpl[3]="hi" #cannot be changed
```

### Sets

```python
#collection oof elements,where you can store elements of diff data types
#unordered collection 
#no duplicates
#no changing of values in sets
st={4,34.5,"hello",False,False} #no duplicates
print(st) #unordered
print(type(st))
print(len(st))
#st[3]="hi" #no changing
```

### Dictionary

```python
# collection of elements, where the elements are stored in key and value pairs
# key and value pairs
# key should be unique
# value can be of any data types
# ordered collection
# can be changed 
# no duplicates
dict={"name":"test","age":45,
      "address":["mangaluru","udupi"],
      "tel":{234576690},
      "tel":{234576690},#no duplicates
      "qualification":{"ug":"bca","pg":"mca"}}
print(dict)
print(type(dict))
print(len(dict))
print(dict["address"])
dict["age"]=56
print(dict)
#print the values : mca,mangaluru
print(dict["qualification"]["pg"])
print(dict["address"][0])
```

### Indexing and Slicing

```python
lst=["apple","banana","cherry","kiwi","mango","blueberry","strawberry","peach","watermelon","orange"]
#try with both positive and negative index
#print the below
#mango
print(lst[4])
print(lst[-6])
#orange
print(lst[9])
print(lst[-1])
#[banana,cherry,kiwi]
print(lst[1:4])
print(lst[-9:-6])
#[apple,cherry,mango,strawberry]
print(lst[0:7:2])
print(lst[-10:-3:2])
#[kiwi,peach]
print(lst[3::4])
print(lst[-7:-2:4])
#reverse all the elements
print(lst[::-1])
#[mango,kiwi,cherry]
print(lst[4:1:-1])
print(lst[-6:-9:-1])
#[orange,peach,blueberry]
print(lst[9:4:-2])
print(lst[-1:-6:-2])
```
