```python
#styling line graphs

a=[1,2,3,4,5]
b=[45,21,65,11,89]
plt.plot(a,b,color='purple',linestyle='-.',linewidth='5',marker='s',markersize='10',markerfacecolor='g',markeredgecolor='r')
plt.show()

#red--r,blue--b,black--k,green--g,yellow--y
#o--circle,s---square,D,d--diamond, H,h--hexagon, <,>,^--triangle, +,*--star
#linestyle--  :,-,--,-.
```

```python
#short method--line color, marker style, line style
a=[1,2,3,4,5]
b=[45,21,65,11,89]
plt.plot(a,b,'go-',markeredgecolor='y')
plt.show()
```

```python
#title, xlabel, ylabel ,figsize ,grid
a=[67,89,91,42,34,12,23]
b=['p1','p2','p3','p4','p5','p6','p7']
plt.figure(figsize=(10,5),dpi=100) #dpi --dot per inch --used for the quality of the image
plt.plot(b,a,'bD-.')
plt.title("Patient age data",color='y')
plt.xlabel("Patient Id",color='magenta')
plt.ylabel("Patient Age",color='magenta')
plt.grid(color='y')
plt.show()
```

```python
#fontdict, saving graph

student_id=['s1','s2','s3','s4','s5']
screen_time=[1,4,3,6,8]
plt.plot(student_id,screen_time)
plt.title("Student Screen Time data",fontdict={'family':'Serif','color':'purple','weight':'bold','size':30})
plt.xlabel("Student Id",fontdict={'family':'Serif','color':'purple','weight':'bold'})
plt.ylabel("Screen time (in Hours)",fontdict={'family':'Serif','color':'purple','weight':'bold'})
plt.savefig('student.png',dpi=100)
plt.show()
```

```python
# time series data, for continuous data---line graph, histogram, scatter plot
#categorical data--bar chart, pie chart, horizontal
```

```python
#bar chart--bar()
subjects=['c','c++','java','python','html']
marks=[78,98,84,88,91]
c=['blue','pink','orange','purple','green']
plt.bar(subjects,marks,color=c)
plt.title("Student Score data",fontdict={'family':'Serif','color':'purple','weight':'bold','size':30})
plt.xlabel("Student Subjects",fontdict={'family':'Serif','color':'purple','weight':'bold'})
plt.ylabel("Screen Marks",fontdict={'family':'Serif','color':'purple','weight':'bold'})
plt.show()
```

```python
subjects=['c','c++','java','python','html']
marks=[78,98,84,88,91]
c=['blue','pink','orange','purple','green']
plt.barh(subjects,marks,color=c)
plt.title("Student Score data",fontdict={'family':'Serif','color':'purple','weight':'bold','size':30})
plt.xlabel("Student Subjects",fontdict={'family':'Serif','color':'purple','weight':'bold'})
plt.ylabel("Screen Marks",fontdict={'family':'Serif','color':'purple','weight':'bold'})
plt.show()
```

```python
#scatter plot--scatter()
age=[67,89,91,42,34,12,23]
patientID=['p1','p2','p3','p4','p5','p6','p7']
c=['blue','pink','orange','purple','green','skyblue','magenta']
plt.scatter(patientID,age,color=c)
plt.title("Age of the patient v/s PatientID",fontdict={'family':'Serif','color':'purple','weight':'bold','size':20})
plt.xlabel("Patient ID",fontdict={'family':'Serif','color':'green','weight':'bold'})
plt.ylabel("Age",fontdict={'family':'Serif','color':'green','weight':'bold'})
plt.show()
```

```python
#histogram--hist()
age=[23,9,61,30,56,78,60,24,55,78,12,35,87,43,28]
a=[0,50,100] #0-50,50-100 -split
plt.hist(age,bins=a,color='g')
plt.show()
```

```python
# pie char--pie()

marks=[78,98,84,88,91]
e=[0.1,0.3,0,0,0]
subjects=['c','c++','java','python','html']
plt.pie(marks,labels=subjects,autopct='%1.f%%',explode=e)
plt.show()
```

```python
x1=[1,12,19,14,25]
y1=[22,44,33,66,88]
x2=[12,14,26,28,30]
y2=[77,55,67,45,88]
plt.plot(x1,y1,'gH:')
plt.plot(x2,y2,'rD--')
plt.title("Advertisement effect on sales")
plt.xlabel("Medium")
plt.ylabel("Sales")
plt.show()
```
