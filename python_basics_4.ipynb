### Indexing and Slicing

```python
lst=["apple","banana","cherry","kiwi","mango","blueberry","strawberry","peach","watermelon","orange"]
#try with both positive and negative index
#print the below
#mango
print(lst[4])
print(lst[-6])
#orange
print(lst[9])
print(lst[-1])
#[banana,cherry,kiwi]
print(lst[1:4])
print(lst[-9:-6])
#[apple,cherry,mango,strawberry]
print(lst[0:7:2])
print(lst[-10:-3:2])
#[kiwi,peach]
print(lst[3::4])
print(lst[-7:-2:4])
#reverse all the elements
print(lst[::-1])
#[mango,kiwi,cherry]
print(lst[4:1:-1])
print(lst[-6:-9:-1])
#[orange,peach,blueberry]
print(lst[9:4:-2])
print(lst[-1:-6:-2])
```

### Tuple Methods

```python
#count(x)	Returns the number of times x appears in the tuple.
my_tuple = (1, 2, 3, 2, 4, 2)
print(my_tuple.count(2))
#index(x)	Returns the index of the first item with value x. Raises ValueError if not found.
print(my_tuple.index(3)) 
```

```python
#Built-in Functions for Tuples
#Returns the number of items.
print(len(my_tuple))

#Returns the largest item.
print(max(my_tuple))

#Returns the smallest item.
print(min(my_tuple))

#Sums all numbers in the tuple.
print(sum(my_tuple))

#Returns a new sorted list from the tuple (does not modify original).
print(sorted(my_tuple))

#Converts a list or string into a tuple.
lst=[1,2,3,4,5]
tup_num=tuple(lst)
print(tup_num)

#Returns True if any element is true.
print(any((0, False, 1)))

#Returns True if all elements are true.
print(all((1, True, "Hi")))
print(all((1, False, "Hi")))
```

### Sets and Dictionary Methods

```python
set_num1={9,8,7,0,2,1,4}
#add(x) Adds an element x to the set.
set_num1.add(3)
print(set_num1)

#remove(x) Removes x. Raises KeyError if not found.
set_num1.remove(8)
print(set_num1)

#discard(x)	Removes x. Does not raise error if missing.
set_num1.discard(5)
print(set_num1)

#pop()	Removes and returns a random element.
set_num1.pop()
print(set_num1)

set_num2={10,20,9}
#union(other)	Returns a new set with elements from both sets.
print(set_num1.union(set_num2))

#intersection(other)	Returns a new set with elements common to both.
print(set_num1.intersection(set_num2))

#difference(other)	Returns elements in the first set but not the second.
print(set_num1.difference(set_num2))

#symmetric_difference(other)	Returns elements in either set, but not both.
print(set_num1.symmetric_difference(set_num2))

#clear()	Removes all elements.
print(set_num1.clear())
```

### Built-in Functions for Sets

```python
A = {1,4,3,2}
# len(set)	Returns the number of unique items.
print(len(A))

#max(set) / min(set)	Largest / Smallest item.
print(max(A))
print(min(A))

#sum(set)	Sum of all elements.
print(sum(A))

#sorted(set)	Returns a sorted list of the set's elements.
print(sorted(A))

#set(iterable)	Converts a list/tuple to a set (removes duplicates).
lst=[5,6,8]
stt=set(lst)
print(stt)
```

### Dictionary Methods

```python
student_scores = {'Alice': 85, 'Bob': 92, 'Charlie': 78, 'John':34}
#keys()	Returns a view object of all keys.
print(student_scores.keys())

#values()	Returns a view object of all values.
print(student_scores.values())

#items()	Returns a view object of all (key, value) pairs.
print(student_scores.items())

#get(key, [default])	Returns value for key. Returns None (or default) if missing.
print(student_scores.get("Alice",'not found'))

#update(other)	Updates the dictionary with pairs from another dict or iterable.
student_scores.update({'Charlie':87})
print(student_scores)

#pop(key)	Removes key and returns its value.
print(student_scores.pop('Alice'))

#popitem()	Removes and returns the last inserted (key, value) pair.
last_item = student_scores.popitem() 
print(f"Popped Last Item: {last_item}")
print(f"Dict after popitem: {student_scores}")

#setdefault(key, default)	Returns value of key if present; otherwise inserts key with default value.
val = student_scores.setdefault('Eve', 70) # 'Eve' didn't exist, so she is added
val_existing = student_scores.setdefault('Bob', 0) # 'Bob' exists, so 0 is IGNORED
print(f"After setdefault('Eve'): {student_scores}")
print(f"Result of setdefault('Bob'): {val_existing}\n")
```

```python
scores = {'Charlie': 78, 'Bob': 92,'Alice': 85, 'John':34}

#len(dict) Returns the number of key-value pairs.
print(len(scores))

#sorted(dict) Returns a sorted list of the KEYS.
print(sorted(scores))

#dict(iterable) Creates a dictionary from a list of (key, value) tuples.
print(dict(scores.items()))

#list(dict) Returns a list of the keys.
print(list(scores))
```
