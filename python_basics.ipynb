### Sets and Dictionary Methods

```python
set_num1={9,8,7,0,2,1,4}
#add(x) Adds an element x to the set.
set_num1.add(3)
print(set_num1)

#remove(x) Removes x. Raises KeyError if not found.
set_num1.remove(8)
print(set_num1)

#discard(x)	Removes x. Does not raise error if missing.
set_num1.discard(5)
print(set_num1)

#pop()	Removes and returns a random element.
set_num1.pop()
print(set_num1)

set_num2={10,20,9}
#union(other)	Returns a new set with elements from both sets.
print(set_num1.union(set_num2))

#intersection(other)	Returns a new set with elements common to both.
print(set_num1.intersection(set_num2))

#difference(other)	Returns elements in the first set but not the second.
print(set_num1.difference(set_num2))

#symmetric_difference(other)	Returns elements in either set, but not both.
print(set_num1.symmetric_difference(set_num2))

#clear()	Removes all elements.
print(set_num1.clear())
```

### Dictionary Methods

```python
student_scores = {'Alice': 85, 'Bob': 92, 'Charlie': 78, 'John':34}
#keys()	Returns a view object of all keys.
print(student_scores.keys())

#values()	Returns a view object of all values.
print(student_scores.values())

#items()	Returns a view object of all (key, value) pairs.
print(student_scores.items())

#get(key, [default])	Returns value for key. Returns None (or default) if missing.
print(student_scores.get("Alice",'not found'))

#update(other)	Updates the dictionary with pairs from another dict or iterable.
student_scores.update({'Charlie':87})
print(student_scores)

#pop(key)	Removes key and returns its value.
print(student_scores.pop('Alice'))

#popitem()	Removes and returns the last inserted (key, value) pair.
last_item = student_scores.popitem() 
print(f"Popped Last Item: {last_item}")
print(f"Dict after popitem: {student_scores}")

#setdefault(key, default)	Returns value of key if present; otherwise inserts key with default value.
val = student_scores.setdefault('Eve', 70) # 'Eve' didn't exist, so she is added
val_existing = student_scores.setdefault('Bob', 0) # 'Bob' exists, so 0 is IGNORED
print(f"After setdefault('Eve'): {student_scores}")
print(f"Result of setdefault('Bob'): {val_existing}\n")
```

```python
#file handling using function concept
def file_create(file):
    try:
        s=open(file,"x")
    except:
        print("File already created")
    else:
        print("File is created")
    finally:
        s.close()
def file_write(file,content):
    try:
        s=open(file,"w")
        s.write(content)
    except:
        print("Error in writing content")
    else:
        print("Content written successfully")
    finally:
        s.close()
def file_append(file,content):
    try:
        s=open(file,"a")
        s.write(content)
    except:
        print("Error in appending content")
    else:
        print("Content appended successfully")
    finally:
        s.close()
def file_read(file):
    try:
        s=open(file,"r")
        print(s.read())
    except:
        print("Error in reading file content")
    else:
        print("successfull")
    finally:
        s.close()
def file_delete(file):
    try:
        import os
        os.remove(file)
    except:
        print("file doesnot exist")
    else:
        print("file removed successfully")
while True:
    print("1 Create file")
    print("2 write content to file")
    print("3 append content to file")
    print("4 read content from file")
    print("5 delete file")
    print("6 Exit")
    ch=int(input("enter your choice: "))
    if ch==1:
        a=input("enter the file name:")
        file_create(a)
    elif ch==2:
        a=input("Enter the file name:")
        y=input("Enter the content:")
        file_write(a,y)
    elif ch==3:
        a=input("Enter the file name:")
        y=input("Enter the content:")
        file_append(a,y)
    elif ch==4:
        a=input("Enter the file name:")
        file_read(a)
    elif ch==5:
        a=input("Enter the file name:")
        file_delete(a)
    elif ch==6:
        print("Exiting the system....")
        break
    else:
        print("Invalid choice,Enter between(1-6)")
```

```python
import os
print("Welcome to File Management System")
class FileManager:
    def create(self, filename):
        try:
            with open(filename, "x"):
                pass
            print("File created successfully")
        except FileExistsError:
            print("File already exists")
    def write(self, filename, content):
        try:
            with open(filename, "w") as f:
                f.write(content)
            print("Content written successfully")
        except:
            print("Error in writing content")
    def append(self, filename, content):
        try:
            with open(filename, "a") as f:
                f.write("\n" + content)
            print("Content appended successfully")
        except:
            print("Error in appending content")

    def read(self, filename):
        try:
            with open(filename, "r") as f:
                print("\n--- File Content ---")
                print(f.read())
        except:
            print("Error in reading file")

    def delete(self, filename):
        try:
            os.remove(filename)
            print("File deleted successfully")
        except:
            print("File does not exist")
fm = FileManager()
while True:
    try:
        print("\n-- Menu --")
        print("1.-- Create File --")
        print("2.-- Write Content --")
        print("3.-- Append Content --")
        print("4.-- Read File --")
        print("5.-- Delete File --")
        print("6.-- Exit --")

        ch = int(input("Enter your choice: "))

        if ch == 1:
            name = input("Enter file name: ")
            fm.create(name)

        elif ch == 2:
            name = input("Enter file name: ")
            data = input("Enter content: ")
            fm.write(name, data)

        elif ch == 3:
            name = input("Enter file name: ")
            data = input("Enter content: ")
            fm.append(name, data)

        elif ch == 4:
            name = input("Enter file name: ")
            fm.read(name)

        elif ch == 5:
            name = input("Enter file name: ")
            fm.delete(name)

        elif ch == 6:
            print("Exiting the system...")
            break

        else:
            print("Invalid choice, Enter between (1-6)")

    except ValueError:
        print("Please enter numbers only")
```

```python
print("Welcome to ATM System")


class ATM:

    def __init__(self):
        self.pin = int(input("Set your PIN: "))
        self.balance = float(input("Enter initial balance: "))

    def credit(self):
        amount = float(input("Enter amount: "))
        if amount > 0:
            self.balance += amount
            print("Balance:", self.balance)
        else:
            print("Invalid amount")

    def debit(self):
        amount = float(input("Enter amount: "))
        if amount > 0:
            if amount <= self.balance:
                self.balance -= amount
                print("Balance:", self.balance)
            else:
                print("Insufficient balance")
        else:
            print("Invalid amount")

    def show_balance(self):
        print("Balance:", self.balance)

    def start(self):
        user_pin = int(input("Enter PIN to login: "))

        if user_pin == self.pin:

            while True:
                print("\n1 Credit\n2 Debit\n3 Balance\n4 Exit")
                ch = int(input("Enter choice: "))

                if ch == 1:
                    self.credit()
                elif ch == 2:
                    self.debit()
                elif ch == 3:
                    self.show_balance()
                elif ch == 4:
                    print("Exiting...")
                    break
                else:
                    print("Invalid choice")

        else:
            print("Incorrect PIN... Program closed")


atm = ATM()
atm.start()
```
