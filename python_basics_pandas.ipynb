```python
#data preprocessing
#checking for unique values
s=pd.Series([2,5,7,2,5,8,9,2,6,7,12])
print(s.unique())

#value counts--prints the count of unique values
print(s.value_counts())
```

```python
#data cleaning--handling null values
a=pd.Series([1,np.nan,2,3,5,np.nan,7,3,5,np.nan,11,8])
print(a)
print(a.value_counts())
#checking null values
print(a.isnull())
print(a.isnull().sum())
print(a.isna().sum())
```

```python
#filling null values
#fillna(),bfill(),ffill(),mean(),median(),mode()
#fillna()
a=pd.Series([1,np.nan,2,3,5,np.nan,7,3,5,np.nan,11,8])
print(a.fillna(5))  #int,float anything is alllowed
```

```python
#bfill---fills the null value with next value
print(a.bfill())
print(a.isnull().sum())
```

```python
#ffill()--fills null value with previous value
print(a.ffill())
print(a.isnull().sum())
```

```python
#median()
x=a.mean()
print(x)
print(a.fillna(x))
print(a.isnull().sum())
#or----print(a.fillna(a.mean()))
```

```python
#mode()
x=a.median()
print(x)
print(a.fillna(x))
print(a.isnull().sum())
#or----print(a.fillna(a.median()))
```

```python
#mean()
k=pd.Series([1,np.nan,2,3,5,np.nan,7,3,5,np.nan,11,8])
p=k.mode()
print(p)
print(k.fillna(p[1]))
#print(a.isnull().sum())
#or----print(a.fillna(a.mode()))
```

```python
# Dataset:
# marks = [78, 56, 89, 45, 67, 90, 34, 88, 76, 54, 92, 48, 60, 73, 85]
# a. Create a Series with index as student IDs: S1 to S15
# b. Find students scoring:
# a. Above 80
# b. Between 50 and 75
# c. Count how many students scored below 50
# d. Increase marks by 5 for students scoring below 50
# e. Replace marks above 90 with 95
```

```python
marks = pd.Series([78, 56, 89, 45, 67, 90, 34, 88, 76, 54, 92, 48, 60, 73, 85],index=['S1','S2','S3','S4','S5','S6','S7','S8','S9','S10','S11','S12','S13','S14','S15'])
print(marks[marks>80])
print(marks[(marks>=50) & (marks<=75)])
count=(marks<50).sum()
print(count)
marks[marks < 50]+=5
marks[marks > 90]=95
print(marks)
```

```python
# Dataset:
# temps = [32, 35, None, 40, 38, None, 36, 37, None, 39, 41, None]
# a. Convert to Series
# b. Count null values
# c. Fill nulls with:
# a. Mean
# b. Forward fill
# c. Backward fill
# d. Find days where temperature > 38
# e. Replace temperatures below 33 with 33
```

```python
temps =pd.Series([32, 35, None, 40, 38, None, 36, 37, None, 39, 41, None])
print(temps.isnull().sum())
x=temps.mean()
print(x)
print(temps.fillna(x))
print(temps.bfill())
print(temps.ffill())
print(temps[temps>38])
temps[temps<33]=33
print(temps)
```

```python
# Create this DataFrame:
# data = {
# "roll_no": [1,2,3,4,5,6,7,8],
# "name": ["Amit","Neha","Rahul","Sneha","Vikram","Anjali","Kiran","Pooja"],
# "gender": ["M","F","M","F","M","F","M","F"],
# "marks": [78, 85, 67, 92, 45, 88, 55, 73],
# "grade": ["B","A","C","A","D","B","C","B"]
# }
# Tasks:
# 1. Print students scoring above 80
# 2. Print female students scoring above 80
# 3. Print male students scoring below 60
# 4. Print name and grade of students scoring between 60 and 90
# 5. Print students whose grade is A OR marks > 90
```

```python
data = {
"roll_no": [1,2,3,4,5,6,7,8],
"name": ["Amit","Neha","Rahul","Sneha","Vikram","Anjali","Kiran","Pooja"],
"gender": ["M","F","M","F","M","F","M","F"],
"marks": [78, 85, 67, 92, 45, 88, 55, 73],
"grade": ["B","A","C","A","D","B","C","B"]
}
df=pd.DataFrame(data)
print(df)
print('--------')
print(df[(df['marks']>80)])
print('--------')
print(df[(df['gender']=='F') & (df['marks']>80)])
print('--------')
print(df[(df['gender']=='M') & (df['marks']<60)])
print('--------')
print(df[(df['marks']>60) & (df['marks']<90)][['name','grade']])
print('--------')
print(df[(df['grade']=='A') | (df['marks']>90)])
```
